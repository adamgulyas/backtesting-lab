```python
import pandas as pd
import yfinance as yf
```

```python
# Get price data from Yahoo! Finance for S&P 500, NASDAQ 100, and RUSSELL 2000
market_data_df = yf.download(
    '^GSPC ^NDX ^RUT'
)

# # rename columns from tickers to descriptive names
market_data_df.rename(
    columns={
        '^NDX': 'NASDAQ 100',
        '^RUT': 'RUSSELL 2000',
        '^GSPC': 'S&P 500', 
    },
    inplace=True
)

market_data_df.drop(['Adj Close', 'Volume'], axis='columns', inplace=True)
# market_data_df.dropna(inplace=True)
market_data_df.columns = market_data_df.columns.swaplevel(0, 1)
market_data_df.sort_index(axis=1, inplace=True)
market_data_df.head()
```

```python
market_data_df.to_csv('data/markets_ohlc.csv')
```
